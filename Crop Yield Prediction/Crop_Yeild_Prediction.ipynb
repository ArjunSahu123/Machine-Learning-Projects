## Correlation Heatmap Analysis and Insights

The correlation analysis shows that crop yield is strongly and positively influenced by rainfall (0.89), nitrogen (0.88), potassium (0.86), fertilizer (0.83), and phosphorus (0.80), indicating that both water availability and nutrient levels play a critical role in improving productivity. In contrast, temperature has a strong negative correlation with yield (-0.89), suggesting that higher temperatures may reduce output due to heat stress. Temperature also shows negative relationships with rainfall (-0.97) and key nutrients, implying that rising temperatures may indirectly affect yield by reducing water availability and nutrient effectiveness.

The heatmap also highlights strong inter-correlation among fertilizer-related variables, particularly between fertilizer and nitrogen (0.91), indicating potential redundancy and multicollinearity. Rainfall emerges as a key enabling factor, not only directly improving yield but also supporting nutrient absorption. These patterns suggest that irrigation and temperature management could significantly enhance productivity, while careful feature selection or dimensionality reduction may be required for building robust machine learning models.

## Test Train Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(df.drop('Yeild (Q/acre)', axis =1), df['Yeild (Q/acre)'], test_size = 0.2, random_state=42)
```

### Model Building
I will be using the following models to predict the crop yield:

###### Decision Tree Regressor
###### Random Forest Regressor

```python
from sklearn.tree import DecisionTreeRegressor

#Decision Tree Regressor object
dtr = DecisionTreeRegressor()
```

```python
from sklearn.model_selection import GridSearchCV

#parameters for grid search
para = {
    "max_depth": [2,4,6,8],
    "min_samples_split": [2,4,6,8],
    "min_samples_leaf": [2,4,6,8],
    "random_state": [0,42]
}

#Grid search object
grid = GridSearchCV(dtr, para, cv = 5, n_jobs = -1, verbose = 1)

#fitting the grid search object
grid.fit(X_train, Y_train)

#best parameters
print(grid.best_params_)
```

```python
#decision tree regressor with best parameters
dtr = DecisionTreeRegressor(max_depth = 4, min_samples_leaf = 2, min_samples_split = 8, random_state = 0)

#fitting the model
dtr.fit(X_train, Y_train)

#training score
print(dtr.score(X_train, Y_train))

#prediction
d_pred = dtr.predict(X_test)
```

## Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

#Random Forest Regressor object
rfr = RandomForestRegressor()
```

```python
from sklearn.model_selection import GridSearchCV

#parameters for grid search
para = {
    "n_estimators": [100,200,300],
    "max_depth": [2,4,6,8],
    "min_samples_split": [2,4,6,8],
    "min_samples_leaf": [2,4,6,8],
    "random_state": [0,42]
}

#Grid search object
grid = GridSearchCV(rfr, para, cv = 5, n_jobs = -1, verbose = 1)

#fitting the grid search object
grid.fit(X_train, Y_train)

#best parameters
print(grid.best_params_)
```

```python
#Random Forest Regressor with best parameters
rfr = RandomForestRegressor(max_depth = 4, min_samples_leaf = 2, min_samples_split = 6, n_estimators = 100, random_state = 42)

#fitting the model
rfr.fit(X_train, Y_train)

#training score
print(rfr.score(X_train, Y_train))

#prediction
r_pred = rfr.predict(X_test)
```

## Model Evaluation

### Distribution Plot

```python
fig, ax = plt.subplots(1,2,figsize=(15, 5))

#plotting the actual vs predicted values for decision tree regressor
sns.distplot(Y_test, ax = ax[0]).set_title('Decision Tree Regressor')
sns.distplot(d_pred, ax = ax[0])

#plotting the actual vs predicted values for random forest regressor
sns.distplot(Y_test, ax = ax[1]).set_title('Random Forest Regressor')
sns.distplot(r_pred, ax = ax[1])
```

Distribution plots were used to compare the distribution of actual and predicted yield values for both Decision Tree and Random Forest models. The plots combine histograms and KDE curves to visualize how closely the predicted values match the actual data distribution.

The Random Forest model shows a closer alignment between predicted and actual distributions, indicating better performance and generalization. In contrast, the Decision Tree model exhibits slight deviations and less smooth predictions. Overall, Random Forest provides more stable and accurate predictions, as reflected by the improved overlap and smoother distribution.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

print("Decision Tree Regressor")
print("Mean Squared Error: ", mean_squared_error(Y_test, d_pred))
print("Mean Absolute Error: ", mean_absolute_error(Y_test, d_pred))
print("R2 Score: ", r2_score(Y_test, d_pred))
print("\n")
print("Random Forest Regressor")
print("Mean Squared Error: ", mean_squared_error(Y_test, r_pred))
print("Mean Absolute Error: ", mean_absolute_error(Y_test, r_pred))
print("R2 Score: ", r2_score(Y_test, r_pred))
```

## Feature Importance

```python
fig, ax = plt.subplots(1,2,figsize=(15, 5))
fig.subplots_adjust(wspace=0.5)

#Decision Tree Regressor
feature_df = pd.DataFrame({'Features':X_train.columns, 'Importance':dtr.feature_importances_})
feature_df.sort_values(by='Importance', ascending=False, inplace=True)
print(feature_df)
sns.barplot(x = 'Importance', y = 'Features', data = feature_df, ax = ax[0]).set_title('Decision Tree Regressor')

#Random Forest Regressor
feature_df = pd.DataFrame({'Features':X_train.columns, 'Importance':rfr.feature_importances_})
feature_df.sort_values(by='Importance', ascending=False, inplace=True)
sns.barplot(x = 'Importance', y = 'Features', data = feature_df, ax = ax[1]).set_title('Random Forest Regressor')
```
